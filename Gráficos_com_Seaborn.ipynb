### Joinplot

#### Junta histogramas com plots e √© util para ver a rela√ß√£o de duas vari√°veis


*   **Ser√° que quanto mair a conta maior √© o valor da corjeta?**
*   **Ser√° que o tamanho da mesa influ√™ncia na gorjeta?**

```
#Vamos verificar total_bill x tips

sb.jointplot(x='total_bill',y='tip',data=tips)
```

A maioria das compras deu entre 10 e 20 d√≥lares.

```
sb.jointplot(x='size',y='tip',data=tips)
```

O tamanho da mesa n√£o influ√™ncia no valor da gorjeta. Na verdade, com base nos dados analisados, podemos dizer que quantos mais lugares uma mesa possu√≠, isto √©, quando maior o n√∫mero de lugares na mesa, menos gorjeta √© dado. 


### Titanic Dataset:

Na base do titanic, ser√° que existe rela√ß√£o entre a idade e o valor pago no ticket de entrada possuem alguma rela√ß√£o?

* **Ser√° que o valor da passagem est√° relacionado com a idade do passageiro?**

```
sb.jointplot(x='fare', y='age',data=titanic)
```

Como podemos ver no gr√°fico acima, n√£o existe rela√ß√£o alguma entre a idade e o valor da passagem. Vemos tamb√©m que a minoria pagou cerga de 500 d√≥lares, os damais pagaram entre 50 e 150 d√≥lares.
