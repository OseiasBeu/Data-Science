### Joinplot

#### Junta histogramas com plots e é util para ver a relação de duas variáveis


*   **Será que quanto mair a conta maior é o valor da corjeta?**
*   **Será que o tamanho da mesa influência na gorjeta?**

```
#Vamos verificar total_bill x tips

sb.jointplot(x='total_bill',y='tip',data=tips)
```

A maioria das compras deu entre 10 e 20 dólares.

```
sb.jointplot(x='size',y='tip',data=tips)
```

O tamanho da mesa não influência no valor da gorjeta. Na verdade, com base nos dados analisados, podemos dizer que quantos mais lugares uma mesa possuí, isto é, quando maior o número de lugares na mesa, menos gorjeta é dado. 


### Titanic Dataset:

Na base do titanic, será que existe relação entre a idade e o valor pago no ticket de entrada possuem alguma relação?

* **Será que o valor da passagem está relacionado com a idade do passageiro?**

```
sb.jointplot(x='fare', y='age',data=titanic)
```

Como podemos ver no gráfico acima, não existe relação alguma entre a idade e o valor da passagem. Vemos também que a minoria pagou cerga de 500 dólares, os damais pagaram entre 50 e 150 dólares.
